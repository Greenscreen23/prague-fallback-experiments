```python
import matplotlib

matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42

import matplotlib.pyplot as plt
import matplotlib.ticker as tck

L_STICKY = 8
CLASSIC_ECN = L_STICKY + 1
C_STICKY = CLASSIC_ECN + L_STICKY

x = [0, L_STICKY, CLASSIC_ECN, C_STICKY]
y = [0, 0, 1 / 4 + 1 / 16, 1 / 4 + 1 / 16]
```

```python
fig, ax = plt.subplots(layout="constrained", figsize=(9, 4))

ax.plot(x, y, color='black')

ax.grid()

ax.set_xlim(0, 17)

ax.yaxis.set_major_locator(tck.IndexLocator(base=0.05, offset=0))
ax.xaxis.set_major_locator(tck.IndexLocator(base=1, offset=0))

ax.set_xlabel("classic_ecn")
ax.set_ylabel("Minimum Congestion\nWindow Reduction Factor")

ax.axvspan(0, L_STICKY, color='orange', alpha=0.5)
ax.axvspan(CLASSIC_ECN, C_STICKY, color='blue', alpha=0.5)

ax.text(4, 0.175, 'Scalable', ha='center', va='center')
ax.text(13, 0.175, 'Classic', ha='center', va='center')

plt.savefig("images/explain/transition.pdf")
```
